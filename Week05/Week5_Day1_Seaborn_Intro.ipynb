```python
flight_filepath='CSV/flight_delays.csv'
flight_data=pd.read_csv(flight_filepath, index_col='Month')
```

```python
flight_data
```

```python
#Average arrival delay for Spirit Airlines (airline code: NK)
plt.figure(figsize=(10,6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])
plt.ylabel("Arrival delay (in minutes)")
```

```python
plt.figure(figsize=(14,7))
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
```

```python
#Part III: Plotting practices
```

```python
sns.set_style("whitegrid")
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
iris = sns.load_dataset("iris")
```

```python
tips.head()
```

```python
flights.head()
```

```python
iris.head()
```

```python
# A. Scatter plot with regression line

plt.figure(figsize=(10, 6))
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title('Relationship Between Total Bill and Tip', fontsize=16)
```

```python
# B. Categorical scatter plot (stripplot)

plt.figure(figsize=(10, 6))
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
plt.title('Distribution of Total Bills by Day', fontsize=16)
```

```python
# C. Box plot

plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Box Plot of Total Bills by Day', fontsize=16)
```

```python
# D. Heatmap

# Pivot the table to have months as rows, years as columns
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")
plt.figure(figsize=(12, 8))

# Create heatmap with annotated values
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu") 
plt.title('Monthly Number of Flight Passengers (1949-1960)', fontsize=16)
```

```python
# E. Pairplot

# Pairplot colored by species
sns.pairplot(iris, hue="species") 
plt.suptitle('Feature Relationships in Iris Dataset', y=1.02, fontsize=16)
```

```python
# F. Violin Plot

# Violin plot showing bill distribution by day and gender
plt.figure(figsize=(12, 8))
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True) 
plt.title('Distribution of Bills by Day and Gender', fontsize=16)
```

```python
# G. Joint Plot

# Hexbin joint plot between total bill and tip
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") 
plt.suptitle('Joint Distribution of Total Bill and Tip', y=1.02, fontsize=16)
```

```python
# H. FaceGrid

# Create FacetGrid by time and sex
g = sns.FacetGrid(tips, col="time", row="sex", height=4)

# Map a scatterplot to the grid
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()  # Add a legend
```
