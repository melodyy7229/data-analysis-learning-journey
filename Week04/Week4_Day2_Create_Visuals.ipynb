```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
```

```python
# Create Sample Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 55, 15]
```

```python
# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.title('Basic Bar Chart', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.show()
```

```python
#A bar chart is best used to compare values across different categories, such as app downloads by genre. 
#It works well when your category names are short and the focus is on the size differences between groups.
```

```python
# Horizontal Bar Chart
plt.figure(figsize=(10, 6))
plt.barh(categories, values, color='salmon')
plt.title('Horizontal Bar Chart', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Category', fontsize=14)
plt.show()
```

```python
#A horizontal bar chart serves the same purpose as a vertical bar chart but is better when category labels are long or when there are many categories to display. 
#Itâ€™s easier to read because the labels are aligned horizontally.
```

```python
# Generate 50 random values between 0 and 1
x = np.random.rand(50)
y = np.random.rand(50)

# Random colors and sizes for each point
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7)
plt.title('Scatter Plot', fontsize=16)
plt.xlabel('X Axis', fontsize=14)
plt.ylabel('Y Axis', fontsize=14)
plt.colorbar(label='Color Value')
plt.show()
```

```python
#A scatter plot is ideal for showing the relationship between two numerical variables, like ratings versus number of reviews. 
#It helps identify patterns, trends, or clusters, and can include color to represent a third variable.
```

```python
# Pie Chart
plt.figure(figsize=(10, 6))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart', fontsize=16)
plt.axis('equal')  # Ensure the pie chart is circular
plt.show()
```

```python
#A pie chart is best for showing proportions of a whole, such as the percentage of apps by content rating. 
#Itâ€™s most effective when there are only a few categories and the values clearly differ.
```

```python
# Multiple Subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Subplot 1: Bar Chart
axs[0, 0].bar(categories, values, color='skyblue')
axs[0, 0].set_title('Bar Chart')

# Subplot 2: Line Chart
axs[0, 1].plot(categories, values, 'r-o', linewidth=2)
axs[0, 1].set_title('Line Chart')

# Subplot 3: Scatter Plot
axs[1, 0].scatter(categories, values, color='green', s=100)
axs[1, 0].set_title('Scatter Plot')

# Subplot 4: Horizontal Bar Chart
axs[1, 1].barh(categories, values, color='salmon')
axs[1, 1].set_title('Horizontal Bar Chart')

plt.tight_layout()
plt.show()
```
