```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('CSV/netflix_titles.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
```

```python
df = df.dropna(subset=['rating', 'duration', 'date_added'])
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='type')
plt.title('Distribution of Content Type on Netflix', fontsize=16)
plt.xlabel('Type', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.tight_layout()
```

```python
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Top 10 Countries Producing Netflix Titles', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of Titles', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
top_ratings = df['rating'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_ratings.index, y=top_ratings.values)
plt.title('Top 10 Most Common Ratings on Netflix', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.tight_layout()
```
